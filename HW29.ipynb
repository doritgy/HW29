```python
import numpy as np 
import pandas as pd
from IPython.display import display
sprt_df = pd.read_csv(r"sport_athletes.csv")
display(sprt_df)
```

```python
#1--a,b,c
sprt_df_max = sprt_df.groupby('Sport Type')[['Weight(kg)', 'Height(cm)', 'Wins']].max().rename(columns={'Weight(kg)': 'max_weight', 'Height(cm)':'max_height', 'Wins':'max_wins'})
display(sprt_df_max)
```

```python
#2
display(sprt_df.groupby('Sport Type')['Weight(kg)'].min().sort_values())
```

```python
#3
display(sprt_df.sort_values(by=['Sport Type', 'Height(cm)'], ascending=[True, False])
.groupby('Sport Type').head(2))
                                                                                       
```

```python
#4a
display(sprt_df.groupby(['Country', 'Sport Type']).max())
#4b
display(sprt_df.groupby(['Country', 'Sport Type']).max().loc[('USA', 'Basketball')])
#4c
display(sprt_df.groupby(['Country', 'Sport Type']).max().xs(level ='Sport Type', key = 'Soccer'))
#4d
display(sprt_df[sprt_df['Sport Type'].isin(['Soccer','Tennis'])].groupby(['Country', 'Sport Type']).max(numeric_only=True).sort_values(by = 'Sport Type'))
#4e
display(sprt_df.groupby(['Country', 'Sport Type']).max(numeric_only=True).swaplevel().sort_index(level=['Sport Type', 'Country']))
#4f
display(sprt_df.groupby(['Country', 'Sport Type']).agg({'Wins': ['min', 'max', 'mean', 'std'], 'Height(cm)': ['min', 'max', 'mean', 'std']}))
```

```python
#5
new_rows = pd.DataFrame({ 'Country': ['Canada', 'USA'], 'Sport Type': ['Hockey', 'Skiing'],
'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'], 'Height(cm)': [183, 172],
'Weight(kg)': [84, 68], 'Wins': [3, 4], 'Olympic Date': ['01/02/2002', '12/02/2010'] }, index = [36,37])
display(pd.concat([sprt_df, new_rows], axis=0)) 
```

```python
#6
import numpy as np 
import pandas as pd
from IPython.display import display
sprt_df = pd.read_csv(r"sport_athletes.csv")
sprt_df['Olympic Date'] = sprt_df['Olympic Date'].str.strip()
sprt_df['Olympic Date'] = pd.to_datetime(sprt_df['Olympic Date'],format='mixed')
display(sprt_df)
sprt_df.set_index('Olympic Date', inplace=True)
display(sprt_df[sprt_df['Wins'] > 0].resample(rule='D').sum())
 
```
